```python
#Import necessary packages
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#Get all the text from the required url
url = requests.get("https://www.iplt20.com/teams/chennai-super-kings/squad/1/ms-dhoni").text
```

```python
#Convert the obtained text using BeatifulSoup 
data = BeautifulSoup(url)
```

```python
#Get all the column names from the Table by looking for table heading: "th"
data_col = data.findAll('th')

for i in range(len(data_col)):
    data_col[i] = data_col[i].text
```

```python
#Since we need on only batting statistics we use only first 11 columns
data_col = data_col[0:-11]
```

```python
#Find the table using find_all function
data.find_all('table')
```

```python
#find the table using attributes, in this case we use "class" 
stats_table = data.find_all("table",attrs={"class":"table table--scroll-on-phablet player-stats-table js-table"})
```

```python
#A function to get the contents of the table
def get_table(tb):
    table = []
    for i in tb:                                             #looking into each table under this table tag:{Batting and bowling}
        tr = i.findAll('tr')                                 #Find all table row in the table
        for x in tr:                                         #Looking into each row of the table
            td = x.findAll('td')                             #Find the contents in each row
            row = [j.text.replace('\n',"").replace("\r","").replace('\t',"") for j in td]    #Clean the text and save it
            table.append(row)                                #Append to the table list
    return table
```

```python
#get the table
table = get_table(stats_table)
table[0] = data_col                                          #Replace first row with colum names     
```

```python
df = pd.DataFrame(table[0:15])                               #Extract only Batting stats
df
```
